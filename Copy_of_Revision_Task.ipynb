```python
# Calculate the probability of randomly selecting a borrowed book from a specific genre.(Business)
borr_book = pd.merge(df1, df2, on='BookID', how='inner')

borr_books = len(borr_book)
Business_borr_books = len(borr_book[borr_book['Genre'] == 'Business'])

probability1 = Business_borr_books / borr_books
print(f'the probability of randomly selecting a borrowed book from business is {probability1:.2%}')
```

```python
# Calculate the probability of randomly selecting a member belonging to a specific membership type (Student)
total_mem = len(df3)

stu_mem = len(df3[df3['MembershipType'] == 'Student'])

probability2 = stu_mem / total_mem
print(f'the probability of randomly selecting a student is {probability2:.2%}')
```

```python
# Calculate the probability of randomly selecting a borrowing record that satisfies two conditions of your choice. (a student and borrow mystery book)
df_full = pd.merge(borr_book, df3, on='MemberID', how='inner')

total_rec = len(df_full)
bor_rec = len(df_full[ (df_full['MembershipType'] == 'Student') & (df_full['Genre'] == 'Mystery') ])

probability3 = bor_rec / total_rec
print(f'the probability of randomly selecting a borrowing record that satisfies the two conditions: a student and borrow mystery book is: {probability3:.2%}')
```

```python
# Find the category with the highest and lowest number of borrowings. / Genre
highest_cat = borr_book['Genre'].value_counts(ascending=False).head(1)

lowest_cat = borr_book['Genre'].value_counts(ascending=True).head(1)

print(highest_cat, lowest_cat)
```

```python
# Find the books that have never been borrowed.
all_books = df1['BookID']
borrowed_books = df2['BookID']

never_borr_books = np.setdiff1d(all_books, borrowed_books)

never_borr_books
```
